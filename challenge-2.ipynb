```python
def img_preprocess(img_path):
    img_np = np.array(Image.open(img_path))
    img_np_pad = 255 - np.zeros([max(img_np.shape[:3]), max(img_np.shape[:3]), 3])
    img_np_pad[:img_np.shape[0], :img_np.shape[1], :3] = img_np
    img_resized_np_pad = cv2.resize(img_np_pad.astype("float"), dsize=(128, 128))
    return img_resized_np_pad
```
